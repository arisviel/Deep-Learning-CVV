```
from mtcnn.mtcnn import MTCNN
import cv2
#Put filename here before cell using!!!!!!!
filename = "7OXf8jPsrA0.jpg"
img = cv2.imread(filename)
detector = MTCNN()
```

```
%%timeit

face_structure = detector.detect_faces(img)
```

```
print(face_structure[0])
left_eye = face_structure[0]["keypoints"]["left_eye"]
right_eye = face_structure[0]["keypoints"]["right_eye"]
radius = 10
color = (255, 0, 0) 
thickness = 2
img2 = cv2.circle(img, left_eye, radius, color, thickness)
img2 = cv2.circle(img, right_eye, radius, color, thickness)
```

```
from IPython.display import Image
cv2.imwrite( "test2.jpeg", img2);
```
